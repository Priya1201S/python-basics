```python
# %%
!pip install langchain_groq langchain_core langchain_community pypdf chromadb sentence_transformers gradio -q

# %%
from langchain_groq.chat_models import ChatGroq
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os
import gradio as gr
import sys # Import sys for checking Python version

# Define your helper functions
def initialize_llm():
    llm = ChatGroq(
        temperature=0,
        # Consider using environment variables for API keys in production
        # groq_api_key=os.environ.get("GROQ_API_KEY"),
        groq_api_key="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", # Replace with your actual key or env var
        model_name="llama-3.3-70b-versatile" # Or your preferred model
    )
    return llm

def create_vector_db():
    # Check if the source directory exists
    source_directory = "/kaggle" # Your directory with PDFs
    if not os.path.exists(source_directory):
        print(f"Source directory not found: {source_directory}. Please check the path.")
        return None

    print(f"Loading documents from {source_directory}...")
    try:
        loader = DirectoryLoader(source_directory, glob='*.pdf', loader_cls=PyPDFLoader)
        documents = loader.load()
        print(f"Loaded {len(documents)} documents.")

        if not documents:
            print("No PDF documents found in the specified directory. Cannot create vector DB.")
            return None

        text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
        texts = text_splitter.split_documents(documents)
        print(f"Split into {len(texts)} text chunks.")

        if not texts:
             print("Document splitting resulted in no text chunks. Cannot create vector DB.")
             return None

        print("Creating embeddings and ChromaDB...")
        embeddings = HuggingFaceBgeEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')

        # Check if embeddings function works
        try:
             # Test with a small piece of text
             test_embedding = embeddings.embed_query("test query")
             print(f"Test embedding created successfully (dim: {len(test_embedding)}).")
        except Exception as embed_error:
             print(f"Error creating test embedding: {embed_error}")
             print("Embeddings model might not be loading correctly.")
             return None


        vector_db_path = "./chroma_db" # Use the relative path
        vector_db = Chroma.from_documents(texts, embeddings, persist_directory=vector_db_path)
        vector_db.persist()

        print(f"ChromaDB created and data saved to {vector_db_path}")
        return vector_db

    except Exception as e:
        print(f"An error occurred during vector DB creation: {e}", file=sys.stderr)
        return None

def setup_qa_chain(vector_db, llm):
    if vector_db is None or llm is None:
        print("Cannot set up QA chain: vector_db or llm is None.")
        return None

    print("Setting up retriever...")
    retriever = vector_db.as_retriever()

    prompt_templates = """ You are a compassionate mental health chatbot. Respond thoughtfully to the following question:
    {context}
    User: {question}
    Chatbot: """
    PROMPT = PromptTemplate(template=prompt_templates, input_variables=['context', 'question'])

    print("Initializing RetrievalQA chain...")
    try:
        qa_chain = RetrievalQA.from_chain_type(
            llm=llm,
            chain_type="stuff",
            retriever=retriever,
            chain_type_kwargs={"prompt": PROMPT}
        )
        print("RetrievalQA chain initialized.")
        return qa_chain
    except Exception as e:
        print(f"Error setting up RetrievalQA chain: {e}")
        return None


# --- Initialization before Gradio App launch ---

print("--- Initializing Chatbot Components ---")
llm = None
vector_db = None
qa_chain = None # Initialize qa_chain in the global scope

try:
    print("Initializing LLM...")
    llm = initialize_llm()
    print("LLM initialized.")
except Exception as e:
    print(f"Error initializing LLM: {e}")


db_path = "/chroma_db" # Use the relative path consistently

if llm is not None: # Only proceed if LLM initialized
    if not os.path.exists(db_path):
        print(f"Chroma DB not found at {db_path}. Attempting to create...")
        vector_db = create_vector_db()
    else:
        print(f"Chroma DB found at {db_path}. Loading...")
        try:
            embeddings = HuggingFaceBgeEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
            vector_db = Chroma(persist_directory=db_path, embedding_function=embeddings)
            print("Chroma DB loaded successfully.")
        except Exception as e:
            print(f"Error loading Chroma DB: {e}")
            # Attempt to recreate if loading fails, as the DB might be corrupted or incompatible
            print("Attempting to recreate Chroma DB...")
            try:
                # Need documents to recreate, assuming they are still in /kaggle
                print("Loading documents to recreate DB...")
                loader = DirectoryLoader("/kaggle", glob='*.pdf', loader_cls=PyPDFLoader)
                documents = loader.load()
                if documents:
                    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
                    texts = text_splitter.split_documents(documents)
                    if texts:
                        vector_db = Chroma.from_documents(texts, embeddings, persist_directory=db_path)
                        vector_db.persist()
                        print("Chroma DB recreated successfully.")
                    else:
                         print("Recreation failed: No text chunks from documents.")
                         vector_db = None # Ensure vector_db is None if recreation fails
                else:
                    print("Recreation failed: No documents found in source directory.")
                    vector_db = None # Ensure vector_db is None if recreation fails
            except Exception as recreate_error:
                 print(f"Error during Chroma DB recreation: {recreate_error}")
                 vector_db = None # Ensure vector_db is None if recreation fails


# Set up the QA chain only if both vector_db and llm are valid
if vector_db is not None and llm is not None:
    qa_chain = setup_qa_chain(vector_db, llm)
    if qa_chain is not None:
        print("QA chain successfully initialized and ready.")
    else:
        print("Failed to initialize QA chain.")
else:
    print("QA chain cannot be initialized due to missing LLM or Vector DB.")

print("--- Chatbot Component Initialization Complete ---")

# --- Gradio Chatbot Response Function ---
# This function will now access the globally defined qa_chain

def chatbot_response(user_input, history=[]):
    # This function runs in a context where the global variables defined above should be available

    if not user_input.strip():
        return "Please provide a valid input."

    # Check if qa_chain was successfully set up before trying to use it
    # This check is crucial as qa_chain could be None if initialization failed
    if qa_chain is None:
        print("Chatbot is not initialized (qa_chain is None). Returning error message.")
        return "Chatbot is not initialized. Knowledge base could not be loaded or created. Please check the server logs."

    try:
        print(f"Processing user input: {user_input}")
        # Use the globally initialized qa_chain
        response = qa_chain.run(user_input)
        print(f"Generated response (partial): {response[:100]}...") # Print partial response for debugging
        return response
    except Exception as e:
        print(f"Error during QA chain execution: {e}", file=sys.stderr)
        return "An error occurred while processing your request."

# --- Launch Gradio App ---

print("\n--- Launching Gradio App ---")
# Use share=True if you need a public URL
# Use debug=True for detailed server logs in the console
with gr.Blocks(theme='Respair/Shiki@1.2.1') as app:
    gr.Markdown("# 🧠 Mental Health Chatbot 🤖")
    gr.Markdown("A compassionate chatbot designed to assist with mental well-being. Please note: For serious concerns, contact a professional.")

    # ChatInterface will call the chatbot_response function
    chatbot = gr.ChatInterface(fn=chatbot_response, title="Mental Health Chatbot")

    gr.Markdown("This chatbot provides general support. For urgent issues, seek help from licensed professionals.")

# Launch the app
# Consider using a specific port if needed, e.g., app.launch(port=7860)
app.launch()
```
